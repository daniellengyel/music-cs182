# Get user factors

```python
with open("./data/song_factors.pkl", "rb") as f:
    song_id_factors = pickle.load(f)
```

```python
song_ids = list(song_id_factors.keys())
song_factors = np.array(list(song_id_factors.values()))
```

# Get song_id to song_name mapping


```python
# for song_id in data['song'].values:
# let's redo all this work in SQLite in a few seconds
t1 = time.time()
# connect to database to get the metadata from MSD
conn = sqlite3.connect(os.path.join(msd_subset_addf_path,
                                    'subset_track_metadata.db'))

q = "SELECT DISTINCT song_id, title, artist_name FROM songs"
#     q += " WHERE song_id='" + song_id + "'"
res = conn.execute(q)
res = res.fetchall()
conn.close()
t2 = time.time()
# song_title_dict = {r[0]: "artist: " + r[2] + ", song: " + r[1] for r in res}
artists = [r[2] for r in res]
```

```python
filtered_song_ids = np.array(["artist: " + r[2] + ", song: " + r[1] for r in res])
```

# Create t-SNE

```python
X_embedded = TSNE(n_components=2).fit_transform(song_factors)
X_embedded.shape
```

```python
# create a new plot with a title and axis labels
trace = go.Scatter(x=X_embedded[:,0], y=X_embedded[:,1], mode='markers',
    marker=dict(
        size=5,
        color = colors, #set color equal to a variable
        colorscale='Viridis',
        showscale=True
    ), text=np.array(filtered_song_ids))
data = [trace]
plotly.offline.iplot(data, filename="tsne-scatter")
```
